##

### Checking for missing values 

```python
data.isnull().sum()
```

```python
mean_fa = data_s.fixed_acidity.loc['mean']
mean_va = data_s.volatile_acidity.loc['mean']
mean_c = data_s.chlorides.loc['mean']
mean_fsd = data_s.free_sulfur_dioxide.loc['mean']
mean_d = data_s.density.loc['mean']
mean_q = data_s.quality.loc['mean']
```

```python
n_data = data.copy()
```

```python
n_data.fixed_acidity = data.fixed_acidity.fillna(mean_fa)
n_data.volatile_acidity = data.volatile_acidity.fillna(mean_va)
n_data.chlorides = data.chlorides.fillna(mean_c)
n_data.free_sulfur_dioxide = data.free_sulfur_dioxide.fillna(mean_fsd)
n_data.density = data.density.fillna(mean_d)
n_data.quality = data.quality.fillna(mean_q)
```

```python
n_data.isnull().sum()
```

### Checking for duplicated values 

```python
n_data.duplicated()
```

```python
n_data =n_data.drop_duplicates()
```

```python
n_data.duplicated()
```

```python
n_data_s =n_data.describe()
n_data_s
```

## Scatter graph

```python
%matplotlib inline
```

```python
plt.title('Fixed acidity over pH', fontdict= {'fontweight': 'bold'})
plt.scatter(x='pH',y='fixed_acidity',data = n_data)
plt.xlabel('pH')
plt.ylabel('Fixed acidity')

plt.legend()
```

### From the graph it can be sayd that those with a lower pH had a higher fixed acidity and those with higher pH had a lower fixed acidity.
### Therefore as the pH of a red wine increases, it's fixed acidity decreases.

## Histogram

```python
plt.hist(n_data.density, bins= np.arange(0.990,1.003,0.001) , rwidth= 0.7)
plt.xticks(bins)

plt.title('Distribution of densities of Red Wine', fontdict= {'fontweight': 'bold'})
plt.xlabel('Density')
plt.ylabel('Red wine')

plt.show()
```

### From the histogram, most red wines possess densities within the range of  0.996 - 0.997 .
### At the same time few red wines possess densities within the range of 0.990 - 0.991 .
